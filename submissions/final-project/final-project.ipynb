# Preprocessing data could be downloaded from https://www.yelp.com/dataset_challenge/dataset

### Get restaurants from the Charlotte, NC area

```python
import json
from pandas import DataFrame
from pandas import Series
import pandas
business_df = []
with open('yelp_academic_dataset_business.json') as f:
    for row in f:
        item = json.loads(row)
        if item['state']=='NC' and 'Restaurants' in item['categories']:
            business_df.append(item)
business_df = DataFrame(business_df)
#remove unnecessary columns
business_df = business_df.drop(['hours','name','neighborhoods','open','type','full_address','city','state','attributes'],axis = 1)
```

### Binarize the column 'categorise' & standardize latitude and longitude

```python
#standardize latitude and longitude
business_df.latitude = (business_df.latitude - business_df.latitude.mean())/business_df.latitude.std()
business_df.longitude = (business_df.longitude - business_df.longitude.mean())/business_df.longitude.std()
cat = list(set(sum(business_df.categories,[])))
cat.remove('Restaurants')
for item in cat:
    business_df[item] = Series([item in stuff for stuff in business_df.categories]).astype(int)
business_df = business_df.drop('categories',axis = 1)
# business_df = business_df.set_index('business_id')
average_review = business_df.review_count.mean()
for i in range(len(business_df)):
# assign 1 if # of reviews great than average # of reviews, else 0. 
    business_df.review_count[i] = (business_df.review_count[i] >= average_review) + 0
        
```

### Get reviews that is about restaurants in Charlotte, NC & binarize review time by months

```python
review_df = []
reviewtime = []
with open('yelp_academic_dataset_review.json') as f:
    for row in f:
        item = json.loads(row)
        if item['business_id'] in list(business_df.business_id):
#             months = [0]*12
#             months[int(item['date'][5:7])-1] = 1
            review_df.append(item)
#             reviewtime.append(months)
review_df = DataFrame(review_df)
reviewtime = DataFrame(reviewtime)
# reviewtime.columns = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
review_df = pandas.concat([review_df,reviewtime],axis = 1)
# review_df = review_df.set_index('user_id')
review_df = review_df.sort_index(by = 'user_id')
review_df = review_df.drop(['date','text','type','review_id','votes','stars'],axis = 1)
```

### Extract active users (with more than 50 reviews)

```python
#drop inactive users (who has less than 50 reviews)
users = review_df.user_id.unique()
freqtable = review_df.user_id.value_counts()
freqUsers = [x for x in users if freqtable.ix[x] >= 50]
mask = review_df.user_id.isin(freqUsers)
review_df = review_df[mask]
```

### Construct input vectors for training Nearest Centroid and SVM

```python
#create properties df for restaurants and concat' it to review_df
col = list(business_df.columns)
bus_df = DataFrame(columns = col)
for item in list(review_df.business_id):
    bus_df = bus_df.append(business_df[business_df.business_id == item])
bus_df = bus_df.set_index(review_df.index)
bus_df = bus_df.drop('business_id',axis=1)
review_df = pandas.concat([review_df,bus_df],axis=1)
```

```python
userGroups = review_df.groupby('user_id')
users = review_df.user_id.unique()
```

### Perform Nearest Centroid - SVM combination scheme for each active user (This part takes time)

```python
from sklearn.neighbors.nearest_centroid import NearestCentroid
from sklearn import svm
import numpy as np
from sklearn import grid_search
performance = []
for i in range(len(userGroups)):
#data preprocessing for each active user    
    dfpos = userGroups.get_group(users[i])
    dfneg = business_df
    mask = dfneg.business_id.isin(dfpos.business_id)
    mask = [not j for j in mask]
    dfneg = dfneg[mask]
    y_pos = Series([1]*len(dfpos))
    y_neg = Series([0]*len(dfneg))
    dfpos = dfpos.drop(['business_id','user_id'],axis = 1)
    dfpos['label']=list(y_pos)
    dfneg = dfneg.drop(['business_id'],axis = 1)
    dfneg['label']=list(y_neg)
    dfneg.columns = dfpos.columns
    x_L = dfpos.drop('label',axis=1).values
    y_L = dfpos.label.values
    x_U = dfneg.drop('label',axis=1).values
    y_U = dfneg.label.values
    trainSize = int(len(dfpos)*0.6)
    testSize = len(dfpos)-trainSize
    x_test = x_L[-testSize:]
    y_test = y_L[-testSize:]
    x_L = x_L[range(trainSize)]
    y_L = y_L[range(trainSize)]
#Use nearest centroid to find the most likely outliers
    clf = NearestCentroid()
    count = 0
    while True:
        clf.fit(np.concatenate((x_L,x_U)),np.concatenate((y_L,y_U)))
        mask = clf.predict(x_U)
        if count == 0:
            Q = x_U[mask==1]
        else:
            Q = np.concatenate((Q,x_U[mask==1]))
        x_U = x_U[mask==0]
        y_U = y_U[mask==0]
        if sum(mask)==0:
            break
        count += 1
#train SVM with outlier and positive samples
    #cross validation using grid search
    svc = svm.SVC(kernel='rbf')
    parameters={'C':np.linspace(1,20,num=20), 'gamma': [0.001,0.01,0.1,1,2,3]}
    clf = grid_search.GridSearchCV(svc, parameters, cv = 3)
    index = np.arange(len(np.concatenate((y_L,y_U))))
    np.random.shuffle(index)
    clf.fit(np.concatenate((x_L,x_U))[index],np.concatenate((y_L,y_U))[index])
    #train SVM with the optimal gamma and cost
    svc = svm.SVC(gamma = clf.best_params_['gamma'], C = clf.best_params_['C'], kernel='rbf')
    svc.fit(np.concatenate((x_L,x_U)),np.concatenate((y_L,y_U)))
    performance.append(svc.score(x_test,y_test))
```
