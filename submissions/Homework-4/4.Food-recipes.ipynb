## Data Parsing and Representation

```python
import csv
allrf = csv.reader(open('allr_recipes.txt','r'),delimiter = '\t')
allrRec = list(allrf)
epicf = csv.reader(open('epic_recipes.txt','r'),delimiter = '\t')
epicRec = list(epicf)
menuf = csv.reader(open('menu_recipes.txt','r'),delimiter = '\t')
menuRec = list(menuf)
totalRec = allrRec + epicRec + menuRec
```

```python
location = []
for item in menuRec:
    if item[0] not in location:
        location.append(item[0])
```

```python
chineseRec = []
mexicoRec = []
japaneseRec = []
koreanRec = []
asianRec= []
east_asianRec = []
westernRec = []
italianRec = []
for item in menuRec:
    name = item[0]
    if name == 'chinese':
        chineseRec.append(item[1:])
    elif name == 'mexico':
        mexicoRec.append(item[1:])
    elif name == 'japanese':
        japaneseRec.append(item[1:])
    elif name == 'east_asian':
        east_asianRec.append(item[1:])
    elif name == 'asian':
        asianRec.append(item[1:])
    elif name == 'korean':
        koreanRec.append(item[1:])
    elif name == 'italian':
        italianRec.append(item[1:])    
    else:
        westernRec.append(item[1:])    
```

```python
location
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import *
from nltk.stem.porter import *
finalchineseRec = []
finalmexicoRec = []
finaljapaneseRec = []
finalkoreanRec = []
finalasianRec= []
finaleast_asianRec = []
finalwesternRec = []
finalitalianRec = []
stemmer = PorterStemmer()
for recipe in chineseRec:
    finalchineseRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in mexicoRec:
    finalmexicoRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in japaneseRec:
    finaljapaneseRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in koreanRec:
    finalkoreanRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in asianRec:
    finalasianRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in east_asianRec:
    finaleast_asianRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in westernRec:
    finalwesternRec.append(' '.join([stemmer.stem(word) for word in recipe]))
for recipe in italianRec:
    finalitalianRec.append(' '.join([stemmer.stem(word) for word in recipe]))    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', min_df=0.05,max_df=0.9)
vectorsChinese = vectorizer.fit_transform(finalchineseRec)
vectorsJapanese = vectorizer.fit_transform(finaljapaneseRec)
vectorsKorean = vectorizer.fit_transform(finalkoreanRec)
vectorsMexico = vectorizer.fit_transform(finalmexicoRec)
vectorsAsian = vectorizer.fit_transform(finalasianRec)
vectorsEast_asian = vectorizer.fit_transform(finaleast_asianRec)
vectorsWestern = vectorizer.fit_transform(finalwesternRec)
vectorsItalian = vectorizer.fit_transform(finalitalianRec)
```

## Clustering & Visualization Chinese Food

```python
import seaborn as sns
import sklearn.metrics.pairwise as pdist
%matplotlib inline
distanceMatrix = pdist.euclidean_distances(vectorsChinese)
cg = sns.clustermap(distanceMatrix, method="complete", figsize=(13,13),xticklabels=False,z_score=1)
print cg.dendrogram_col.reordered_ind
```

## Clustering & Visualization Japanese Food

```python
distanceMatrix = pdist.euclidean_distances(vectorsJapanese)
cg = sns.clustermap(distanceMatrix, method="complete", figsize=(10,10), xticklabels=False,z_score=1)
print cg.dendrogram_col.reordered_ind
```

## Clustering & Visualization Italian Food

```python
distanceMatrix = pdist.euclidean_distances(vectorsItalian)
cg = sns.clustermap(distanceMatrix, method="complete", figsize=(10,10), xticklabels=False,z_score=1)
print cg.dendrogram_col.reordered_ind
```
